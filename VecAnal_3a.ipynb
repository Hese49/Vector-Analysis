### 3a. Equation of a plane


**Equation of a plane perpendicular to a given vector.**

**Problem.** Find an equation for the plane perpenticular to the vector $\vec{A} = 2\vec{i} + 3\vec{j}  + 6\vec{k}$ and passing trough the terminal point of the vector $\vec{B} = \vec{i} + 5\vec{j}  + 3\vec{k}$. [Sch21/18]

```python
from sympy import *
x, y, z= symbols('x y z')
from sympy.vector import CoordSys3D
C = CoordSys3D('C')              # Määrittelee standardikoordinaatiston

r = x*C.i + y*C.j + z*C.k        # Position vector for points on the plane.

# INPUT
A = 2*C.i + 3*C.j  + 6*C.k       # The plane must be perpendicular to the this vector.
B = C.i + 5*C.j + 3*C.k          # The plane must pass through the vertex of this vector (1,5,3). 

# CALCULATE
r = x*C.i + y*C.j + z*C.k        # Position vector for points on the plane.
PQ = B-r                         # This vector must lie in the plane and
                                 # be perpendicular to the vector A

EQ = Eq(-1*PQ.dot(A),0)          # The plane equation is obtained when the 
                                 # inner product of the vectors PQ and A is set
                                 # to zero. (-1* to get +sign for variable terms)
# OUTPUT
EQ
```

The same equation can also be achieved using SymPy Matrix.

Planes: &nbsp;  $\vec{n} \cdot (P_0-\left<x,y,z\right>) = 0$

```python
P0 = Matrix([1,5,3])
r =  Matrix([x,y,z])
n =  Matrix([2,3,6])
eq = n.dot(P0-r)
print(eq)
eq = Eq(-1*eq,0) 
eq

```
